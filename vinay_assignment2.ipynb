# Querying a Data Frame


```
import numpy as np
import pandas as pd
```

```
dataset = pd.read_csv('/content/train-200907-141856.csv')
dataset.head()
```

```
dataset['Name'].head()
# dataset.Name.head()
```

```
dataset['Age'].min()
```

```
dataset['Age'].max()
```

```
dataset['Age'].mean()
```

Using loc() and iloc() for selecting data 

```
dataset.loc[0:10, 'Name']
```

```
dataset.loc[0:10, 'Name':'Ticket']
```

```
dataset.iloc[0:10, 3:9]
```

### Indexing and Row Selection

```
# Print the name of the oldest passenger travelling on ship
dataset[dataset['Age'] == dataset['Age'].max()]['Name']
```

### Sorting

```
dataset.sort_values(by='Name').head()
```

```
dataset.sort_values(by=['Name', 'Fare'], ascending=[True, False]).head()
```

### Replacing values in column

```
new = {'male':0, 'female':1}
dataset = dataset.replace({'Sex':new})
dataset.head()
```

```
old = {0:'male', 1:'female'}
dataset['Sex'] = dataset['Sex'].map(old)
dataset.head()
```

Using gropby()

```
dataset.groupby(by='Survived')['Age'].describe()
```

```
dataset.groupby(by='Embarked')['Age'].agg(np.mean)
```

### Using Contingency Table for showing relation between variables

```
pd.crosstab(dataset['Survived'], dataset['Sex'], normalize=True)
```

```
dataset.pivot_table(['Age', 'Fare'],['Survived'])
```
