# Data Visualization

```
import numpy as np
import pandas as pd
```

```
dataset = pd.read_csv('/content/train-200907-141856.csv')
dataset.head()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
sns.countplot(x='Survived', data = dataset)
```

```
sns.countplot(x='Survived', hue='Sex', data = dataset)
```

```
sns.lineplot(x='Age', y='Fare', data = dataset)
```

```
dataset.plot.scatter(x='Age', y='Fare')
```

```
dataset.plot.scatter(x='Age', y='Survived')
```

```
dataset['Age'].hist(bins=10)
```

```
sizes= dataset['Survived'].value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels=['Not Survived', 'Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```
